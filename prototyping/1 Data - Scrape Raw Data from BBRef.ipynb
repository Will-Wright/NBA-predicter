```python
# Web scraper utility
#    https://github.com/jaebradley/basketball_reference_web_scraper
# Local directory for web scraper:
#    /usr/local/lib/python3.7/site-packages/basketball_reference_web_scraper
# Note scraper has been modified to return advanced statistics

import os
import csv
from basketball_reference_web_scraper import client
from basketball_reference_web_scraper.data import OutputType
```

```python
# Scrapes raw data for player box scores
#
# DONE FOR SEASONS 2000-2001 to current
#

month_range = [10, 11, 12, 1, 2, 3, 4, 5, 6]
year_range = range(2007, 1998, -1)
file_path = "data_raw/player_box_scores/"

for season_start_year in year_range:
    season_str = str(season_start_year) + "_" + str(season_start_year + 1)
    if not os.path.exists(file_path + season_str):
        os.mkdir(file_path + season_str)

    year = season_start_year

    for month in month_range:
        for day in range(1,32):
            if month == 1 and day == 1:
                year = year + 1

            if day < 10:
                day_str = "0" + str(day)
            else:
                day_str = str(day)
            if month < 10:
                month_str = "0" + str(month)
            else:
                month_str = str(month)
            output_file_path = "./" + file_path + season_str \
                + "/" + str(year) + "_" + month_str + "_" + day_str \
                + "_" + "player_box_scores.csv"

            if not os.path.exists(output_file_path):
                pbs = client.player_box_scores(day=day, month=month, year=year, 
                                               output_type=OutputType.CSV, 
                                               output_file_path=output_file_path)
                print()
                with open(output_file_path) as csv_file:
                    csv_reader = csv.reader(csv_file, delimiter=',')
                    line_count = 0
                    for row in csv_reader:
                        line_count += 1
                if line_count < 2:
                    os.remove(output_file_path)
                    print(str(year) + "_" + month_str + "_" + day_str + ": No games played")
                else:
                    print(str(year) + "_" + month_str + "_" + day_str + ": Game data saved")
            else:
                print(str(year) + "_" + month_str + "_" + day_str + ": Game data already exists")


```

```python
# Scrapes raw data for season schedule
#
# DONE FOR SEASONS 2000-2001 to current
#

month_range = [10, 11, 12, 1, 2, 3, 4, 5, 6]
year_range = range(2019, 1998, -1)
file_path = "data_raw/season_schedule/"

for season_start_year in year_range:
    season_str = str(season_start_year) + "_" + str(season_start_year + 1)

    season_end_year = season_start_year + 1
    
    output_file_path = "./" + file_path \
        + str(season_start_year) + "_" + str(season_end_year) \
        + "_" + "season_schedule.csv"

    pbs = client.season_schedule(season_end_year=season_end_year, 
                                output_type=OutputType.CSV, 
                                output_file_path=output_file_path)
    with open(output_file_path) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        for row in csv_reader:
            line_count += 1
    if line_count < 2:
        os.remove(output_file_path)
        print(str(season_start_year) + ": No games played")
    else:
        print(str(season_start_year) + ": Game data saved")

```
